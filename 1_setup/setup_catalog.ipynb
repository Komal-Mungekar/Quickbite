## Upload source data to a Databricks Volume
In the Databricks UI:
1. Go to **Catalog > quickbite > Create Volume** → name it `source_data`
2. Upload each folder from `data/` into the volume under `raw/`

Expected paths:
- `/Volumes/quickbite/source_data/raw/restaurants/*.csv`
- `/Volumes/quickbite/source_data/raw/customers/*.csv`
- `/Volumes/quickbite/source_data/raw/delivery_agents/*.csv`
- `/Volumes/quickbite/source_data/raw/menu_items/*.csv`
- `/Volumes/quickbite/source_data/raw/date/*.csv`
- `/Volumes/quickbite/source_data/raw/orders/landing/*.csv`

````python
# Verify volume is accessible
dbutils.fs.ls('/Volumes/quickbite/source_data/raw/')
````
