# Silver to Gold: Promoting Dimension Tables
Selecting and renaming columns to produce clean and BI ready dimension tables.

```python
import pyspark.sql.functions as F

catalog_name = 'quickbite'
```

## gld_dim_restaurants

```python
df = spark.table(f"{catalog_name}.silver.slv_restaurants")

gld = df.select(
    F.col("restaurant_id"),
    F.col("name").alias("restaurant_name"),
    F.col("city"),
    F.col("cuisine_type"),
    F.col("rating"),
)

gld.write.format("delta") \
    .mode("overwrite") \
    .option("mergeSchema", "true") \
    .saveAsTable(f"{catalog_name}.gold.gld_dim_restaurants")

print("gld_dim_restaurants:", spark.table(f"{catalog_name}.gold.gld_dim_restaurants").count(), "rows")
```

## gld_dim_customers

```python
df = spark.table(f"{catalog_name}.silver.slv_customers")

gld = df.select("customer_id","name","phone","city","signup_date")

gld.write.format("delta") \
    .mode("overwrite") \
    .option("mergeSchema", "true") \
    .saveAsTable(f"{catalog_name}.gold.gld_dim_customers")

print("gld_dim_customers:", spark.table(f"{catalog_name}.gold.gld_dim_customers").count(), "rows")
```

## gld_dim_delivery_agents

```python
df = spark.table(f"{catalog_name}.silver.slv_delivery_agents")

gld = df.select(
    F.col("agent_id"),
    F.col("name").alias("agent_name"),
    F.col("vehicle_type"),
    F.col("city"),
)

gld.write.format("delta") \
    .mode("overwrite") \
    .option("mergeSchema", "true") \
    .saveAsTable(f"{catalog_name}.gold.gld_dim_delivery_agents")

print("gld_dim_delivery_agents:", spark.table(f"{catalog_name}.gold.gld_dim_delivery_agents").count(), "rows")
```
